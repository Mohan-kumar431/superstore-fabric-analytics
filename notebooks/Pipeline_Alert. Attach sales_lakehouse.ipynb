```python
from pyspark.sql import Row
from datetime import datetime

# Create an alert record
alert_row = spark.createDataFrame([
    Row(
        pipeline_name="Load_Superstore_Pipeline",
        failure_time=datetime.now(),
        status="FAILED",
        message="Pipeline failed -- check Monitoring Hub for activity-level details"
    )
])

# Append the alert to the Delta table
alert_row.write \
    .mode("append") \
    .format("delta") \
    .saveAsTable("pipeline_alert_log")

print(f"Alert logged at {datetime.now()}")

# Display the most recent alerts
spark.table("pipeline_alert_log") \
    .orderBy("failure_time", ascending=False) \
    .show(5, truncate=False)
```
